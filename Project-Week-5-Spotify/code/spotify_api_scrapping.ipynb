# Getting Track Information & Dataframe

```python
from config import *
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= Client_ID, client_secret=Client_Secret))
```

```python
username = "spotify"
```

```python
def get_playlist_info(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id) #gets playlist info
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
list_playlists = []

for playlist_id in playlist_ids:
    list_playlists.append(get_playlist_info(username, playlist_id))
```

```python
def main(playlists):
    
    list_of_df = []

    for  playlist in playlists:   
        
        song_names = []
        song_uri_total = []
        artist_names = []
        audio_features = []
        song_uri = []
        counter = 0
        
        for item in playlist:
            if item["is_local"] == False:
                try: 
                    counter += 1
                    song_names.append(item['track']['name'])
                    song_uri_total.append(item['track']['uri'])
                    song_uri.append(item['track']['uri'])
                    
                    temp_artists = ""
                    for artist in item['track']['artists']:
                        temp_artists += artist['name'] + " "
                        
                    artist_names.append(temp_artists)
                    
                    if counter == 100:
                        audio_features.append(sp.audio_features(song_uri))
                        counter = 0
                        song_uri = []
                        
                except TypeError:
                    continue
    

        audio_features.append(sp.audio_features(song_uri))
        
        audio_features_total = [subitem for item in audio_features for subitem in item]
        
        df = pd.DataFrame({"song_name": song_names, "artist_name": artist_names, "uri": song_uri_total})
        
        df_features = pd.DataFrame(audio_features_total)
        
        spotify_df = df.merge(df_features, on="uri", how="left")
        
        list_of_df.append(spotify_df)
        
    return list_of_df
```

```python
df_list = main(playlists=list_playlists)
```
