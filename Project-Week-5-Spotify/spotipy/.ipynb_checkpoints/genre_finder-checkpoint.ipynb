**Searching for Playlists by most popular genres**

```python
from config import *
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= Client_ID, client_secret=Client_Secret))
```

```python
genres = ["pop", "dance pop", "rap", "rock", "post-teen pop", "pop rap", 
          "latin", "hip hop", "modern rock", "trap"]
```

```python
playlist = []

for genre in genres:
     playlist.append(sp.search(genres, type='playlist', limit=20))
```

```python
song_names = []
song_uri = []
song_uri_total = []
artist_names = []
audio_features = []
counter = 0

for item in playlist:
    
    counter += 1
    
    song_names.append(item['track']['name'])
    song_uri_total.append(item['track']['uri'])
    song_uri.append(item['track']['uri'])
    
    temp_artists = []
    for artist in item['track']['artists']:
        temp_artists.append(artist['name'])
    
    artist_names.append(temp_artists)
    
    if counter == 100:
        audio_features.append(sp.audio_features(song_uri))
        counter = 0
        song_uri = []

audio_features.append(sp.audio_features(song_uri))
```
